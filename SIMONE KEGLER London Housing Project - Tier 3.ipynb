**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt the DataFrame
properties_melt = properties_new1.melt(id_vars= ['Borough', 'ID'], var_name= 'Month', value_name='Average Price')

properties_melt['Year'] = properties_melt['Month'].apply(lambda t: t.year)

properties_melt.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Convert the data type of the 'Average Price' column to float
properties_melt['Average Price'] = properties_melt['Average Price'].astype('float')

# Convert the Month column to datetime
properties_melt['Month'] = pd.to_datetime(properties_melt['Month'])

# Remove the time from the date column
import datetime as dt
properties_melt['Month'] = properties_melt['Month'].dt.date

# Check the data types of the DataFrame
properties_melt.dtypes
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Create a function that calculates the price ratio of 2003 to 2023, as a full year of 2024 data is not yet available 
def create_ratio(d):
    y2003 = float(d['Average Price'][d['Year']==2003].iloc[0])
    y2023 = float(d['Average Price'][d['Year']==2023].iloc[0])
    ratio1 = [y2023/y2003]
    return ratio1
```

```python
# Create an empty dictionary to store ratios for each unique borough
ratio1 = {}

# Use for loop to iterate over each borough and calculate the price ratios
for b in dfg['Borough'].unique():
    # Parameterize the borough name
    borough = dfg[dfg['Borough'] == b]
    # Create a new entry in the ratio dictionary whose value is the result of calling the create_price_ratio function with the borough name as the key
    ratio1[b] = create_ratio(borough)
    
print(ratio1)
```

```python
# Test create_price_ratio function
create_ratio(dfg[dfg['Borough']=='Bexley'])
```

```python
# Convert the ratios dictionary to a DataFrame
df_ratios = pd.DataFrame(ratio1)

# Transpose the DataFrame
df_ratios_T = df_ratios.T

# Reset the index
df_ratios_T = df_ratios_T.reset_index()

# Rename the columns
df_ratios_T.rename(columns = {'index':'Borough', 0:'2023'}, inplace=True)

df_ratios_T
```

```python
# Sort 2023 ratios in descending order
ratios_sorted = df_ratios_T.sort_values(by='2023',ascending=False)
print(ratios_sorted)
```

```python
# Sort the DataFrame in descending order
top15 = df_ratios_T.sort_values('2023',ascending=False).head(15)

# Plot the top 15 boroughs
top15.plot(kind='bar', x='Borough', y='2023')
plt.ylabel('Price Ratio')
plt.xlabel('Borough Name')
plt.title('Top 15 Boroughs Price Ratio')
plt.show()
```
